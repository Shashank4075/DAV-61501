## 2.  Univariate Analysis.

```python
def box_dist_plot(data, column):
    
    """
    
    This function is to plot box plot and distribution plot for a given column, 
    column's median value, with count and percentage of null values. 
    
    Parameters :-
        data : Dataframe         # Dataframe contains data.
        column : str             # Column name which is to be ploted.
    
    """
    
    plt.style.use('ggplot')
    
    plt.figure(figsize=(18, 7))                    # to set the figure size
    
    plt.subplot(121)                               
    sns.boxplot(data[column])                      # to create box plot
    
    plt.subplot(122)
    sns.distplot(data[column])                     # to create distribution plot
    
    plt.show()
    
    print(data[column].describe())
    print('Median:', data[column].median())
    print()
    print('Total Num of null values:', data[column].isnull().sum(),'count' ,',',
          round(data[column].isnull().mean()*100,2), '%')
    
```

---------------------------------------------------------------------------------------
When to use which type of Central Tendency
- Mean   → Data is numerical, symmetric and have no outliers.
- Median → Data is numerical and skewed or having outliers.
- Mode   → Used with categorical data, can be used with numerical data also. 
---------------------------------------------------------------------------------------

### Handeling Outliers

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['ResidualSugar'].quantile(0.25) # Q1 -one forthvalue
Q3 = df['ResidualSugar'].quantile(0.75) # Q3 -three froth value

# Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1

# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
df_cleaned = df[(df['ResidualSugar'] >= lower_bound) & (df['ResidualSugar'] <= upper_bound)]

print(df_cleaned)
```

```python
# To fill the NA values with median.

df['ResidualSugar'].fillna(df['ResidualSugar'].median(), inplace = True)
```
